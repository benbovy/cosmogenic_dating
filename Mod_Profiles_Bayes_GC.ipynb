## 1. A mathemactical model for calculating the nucleide concentration profiles

In this section we show the derived equations of the model that we'll use to calculate nucleide concentration profiles. A more detailled description of the theoretical background behind this model can be found in, e.g., Braucher et al., 2003 [NEED REF DOI LINK]. 

The contribution of a specific particule type $p$ to the nucleide concentration can be written as:

$$P_p,\Lambda_p \rightarrow c_p(z,\epsilon,t,\rho) = \frac{P_p} {\frac{\rho \, \epsilon}{\Lambda_p} + \lambda} \cdot \exp\left[{-\frac{\rho \, z}{\Lambda_p}}\right] \left[1 - \exp\left\{-t \left(\frac{\rho \, \epsilon}{\Lambda} + \lambda \right)\right\}\right]$$

where $z$ (g cm$^{-2}$) is the depth below the surface, $\epsilon$ (g cm$^{-2}$yr$^{-1}$) is the erosion rate, $t$ (yr) is the exposure time, $\rho$ (g cm$^{-3}$) is the soil density, $\lambda$ (yr$^{-1}$) is the radioactive decay constant, $P_p$ (atom g$^{-1}$yr$^{-1}$) is the production rate of the nucleide at the surface due to the particule, and $\Lambda_p$ (g cm$^{-2}$) is the effective apparent attenuation depth for that particule type.

The nucleide concentration is the sum of individual contributions plus inheritance (i.e., a function of the concentration of the nucleide at the initiation of the exposure scenario). For example, for $^{10}Be$, we have:


$$
\begin{align}
C(z,\epsilon,t,\rho) = & P_n,\Lambda_n & \rightarrow & c_n(z,\epsilon,t,\rho) & + \\
& P_{\mu \mathrm{Slow}},\Lambda_{\mu \mathrm{Slow}} & \rightarrow & c_{\mu \mathrm{Slow}}(z,\epsilon,t,\rho) & + \\
& P_{\mu \mathrm{Fast}},\Lambda_{\mu \mathrm{Fast}} & \rightarrow & c_{\mu \mathrm{Fast}}(z,\epsilon,t,\rho) & + \\
& C_{t0} \cdot e^{-\lambda t} & &
\end{align}
$$

where $C_{t0}$ is the number of atoms present at the initiation of the exposure scenario (assumed constant along the concentration profile), and $c_n(z,\epsilon,t,\rho)$, $c_{\mu \mathrm{Slow}}(z,\epsilon,t,\rho)$ and $c_{\mu \mathrm{Fast}}(z,\epsilon,t,\rho)$ are the contributions of neutrons, slow muons and fast muons, respectively.

A Python/Numpy vectorized implementation of this model gives:

```python
def compute_pcontrib(depth, erosion, exposure, density,
                     nucleide, particle):
    """
    Returns the contribution of a specific particle type
    to the nucleide concentration.
    
    See Also
    --------
    compute_Cn
    """
    return (
        particle['prod_rate'] / ((density * erosion / 
                                  particle['damping_depth']) 
                                 + nucleide['rdecay']) *
        np.exp(-1. * density * depth / particle['damping_depth']) *
        (1. - np.exp(-1. * exposure * ((erosion * density / 
                                       particle['damping_depth'])
                                       + nucleide['rdecay'])))
    )


def compute_Cn(depth, erosion, exposure, density,
               inheritance, nucleide, particles):
    """
    Calculate the nucleide concentration(s) for given
    parameters.
    
    Parameters
    ----------
    depth : float or array_like
        depth(s) below the surface [cm]
    erosion : float or array_like
        erosion rate(s) [cm yr-1]
    exposure : float or array_like
        exposure time [yr]
    density : float or array_like
        soil density [g cm-3]
    inheritance : float or array_like
        concentration of the nucleide at the
        initiation of the exposure scenario
        [atoms g-1]
    nucleide : dict
        nucleide parameters
    particles : [dict, dict, ...]
        parameters related to each particule type
        that contribute to the nucleide production
    
    Returns
    -------
    float or array-like (broadcasted)
        the nucleide concentration(s) [atoms g-1]
    
    Notes
    -----
    if arrays are given for several arguments, they must
    have the same shape or must be at least be broadcastable.
    
    `nucleide` must have the following key(s):
        'rdecay': radioactive decay [yr-1]
    
    each item in `particles` must have the following keys:
        'prod_rate': surface production rate [atoms g-1 yr-1]
        'damping_depth': effective apparent attenuation depth
                         [g cm-2]
    
    """
    return (
        np.sum((compute_pcontrib(depth, erosion, 
                                 exposure, density,
                                 nucleide, p)
                for p in particles),
               axis=0) +
        inheritance * np.exp(-1. * nucleide['rdecay'] * exposure)
    )
```

## 2. Modelling a measured concentration profile using the Bayesian approach

**NOTE**: Here below, we distinguish between:

- the *mathematical model*: the model defined by the equations in the previous section, which has parameters that are assumed to be well-known and other parameters that we want to fit with data
- a *data model*: the mathematical model with a given value for each of its parameters

Our problem is an inverse problem: we have measured cosmogenic nucleide concentrations along a profile and there is a lack of knowledge about the phenomena that caused these concentrations, i.e., exposure time, erosion rate, soil density, inheritance... In addition, we can rely on a mathematical model to predict nucleide concentrations along a depth profile, given some parameters. Using the observations and the mathematical model, our goal is to improve our knowledge about the phenomena - or more precisely the unknown parameters of the mathematical model - which caused the observed nucleide concentrations.

The bayesian approach can help us. Basically, it allows to calculate a *posterior probability distribution* (PPD), i.e., the probability distribution of data models given the observations, or more intuitively said, the probability distribution of the unknown causes behind the observations.

In our problem, we have (1) a vector formed by the parameters in the mathematical model that we want to fit, noted $\mathbf{m} = (\epsilon, t, \rho, C_{t0})$, which represent any data model in the parameter space ${\cal M}$, and (2) a vector formed by the measured nucleide concentrations along a profile $\mathbf{C^o_z} = (C^o_{z0}, C^o_{z1} ... C^o_{zN})$. Applying the Bayes theorem gives:

$$ P(\mathbf{m} | \mathbf{C^o_z}) = \frac{P(\mathbf{m}) P(\mathbf{C^o_z} | \mathbf{m})}{P(\mathbf{C^o_z})}$$

where $P(\mathbf{m} | \mathbf{C^o_z})$ is the posterior probability distribution (i.e., the probability distribution of  $\mathbf{m}$ given $\mathbf{C^o_z}$), $P(\mathbf{m})$ is the *prior probability distribution* (i.e., our prior knowledge about the parameters of the mathematical model that we want to fit), $P(\mathbf{C^o_z} | \mathbf{m})$ is the probability distribution of the observed concentration profile given the data models, and $P(\mathbf{C^o_z})$ can be treated as a normalizing constant.

$P(\mathbf{C^o_z})$ is in fact the probability of obtaining the measured concentration profile, but it is not easy to calculate. We can still derive it considering that the intergral of $P(\mathbf{m} | \mathbf{C^o_z})$ over $\mathbf{m}$ should be 1.

From a theorical point of view, it is inconsistent to use $P(\mathbf{C^o_z} | \mathbf{m})$ since we don't know the real parameter values (according to the Bayesian theory, we can't find the true values, we can only quantify our degree of belief in the values). Instead we use its "counterpart" $L(\mathbf{m} | \mathbf{C^o_z})$, i.e., the *likelihood function*, which represents the fit of any data model to the observed concentration profile, and which is not a probability density! If we assume that the measured concentration profile values have a known, random error, we can use the Gauss expression:

$$ P(\mathbf{C^o_z} | \mathbf{m}) \sim \, L(\mathbf{m} | \mathbf{C^o_z}) = k \, \exp\left(-\frac{1}{2}[\mathbf{C^o_z} - \mathbf{C^m_z}(\mathbf{m})]^T \, (\mathbf{C^o_V})^{-1} \, [\mathbf{C^o_z} - \mathbf{C^m_z}(\mathbf{m})]  \right) $$

where $\mathbf{C^m_z}(\mathbf{m})$ denotes the concentration profile predicted by the data model, $\mathbf{C^o_V}$ is the covariance matrix related to the measured concentrations along the profile and $k$ is a constant. Since $P(\mathbf{m} | \mathbf{C^o_z})$ can be further normalized by computing its integral over $\mathbf{m}$, it is not important to know the value of $k$.

Our goal is to characterize $P(\mathbf{m} | \mathbf{C^o_z})$ in the multidimensional space ${\cal M}$. The regions which maximize $P(\mathbf{m} | \mathbf{C^o_z})$ is one property of interest. In the absence of prior information (i.e., a uniform prior probability distribution), the data model that maximizes $P(\mathbf{m} | \mathbf{C^o_z})$ would correspond to the best data fit model as it would be obtained using Maximum Likelihood Estimation (MLE).

Here below is a Python/Numpy vectorized implementation that allows to compute $L(\mathbf{m} | \mathbf{C^o_z})$ and $P(\mathbf{m} | \mathbf{C^o_z})$ for one observed concentration profile and one or many profiles calculated by given data model(s). Note that this is a simplified implementation as it assumes independent concentration measurements along the profile. The expression for $L(\mathbf{m} | \mathbf{C^o_z})$ becomes:

$$ L(\mathbf{m} | \mathbf{C^o_z}) = k \, \exp\left(-\frac{\nu}{2} \chi_{\nu}^2(\mathbf{C^o_z}, \mathbf{C^m_z}(\mathbf{m}))\right) $$

where

$$ \chi_{\nu}^2(\mathbf{C^o_z}, \mathbf{C^m_z}(\mathbf{m})) = \frac{1}{\nu} \sum_{i=1}^N \frac{[C^o_{zi} - C^m_{zi}(\mathbf{m})]^2}{(\sigma^o_i)^2} $$

and $\nu$ is the degrees of freedom, i.e., $N - n$ ($N$ is the number of samples along the measured concentration profile and $n$ is the number of parameters to fit). $\sigma^o_i$ is the standard deviation associated to each measured concentration $C^o_{zi}$. 

Note that $\chi_{\nu}^2(\mathbf{C^o_z}, \mathbf{C^m_z}(\mathbf{m}))$, i.e., the reduced or normalized chi-square, measures the *goodness-of-fit* of the data model to the measured concentration profile, and would be the variable to minimize using the *Minimum Chi-square Estimation* (MCE). Compared to MCE or MLE, the Bayesian approach can be viewed as a generalization since it allows to provide prior information about the parameters that we want to fit. From a conceptual point of view, it has also fundamental diffferences ('true' parameter values vs. degree of belief in the values) that are no longer debated here.  

```python
def compute_chi2_reduced(mprofile, oprofile, ostd,
                         deg_freedom):
    """
    Compute the reduced (normalized) chi-square
    given a measured concentration profile and
    a modelled profile.
    
    Parameters
    ----------
    mprofile : 1-d or n-d array_like
        the modelled concentration profile
    oprofile : 1-d array_like
        the measured concentration profile
    ostd : 1-d array_like
        the standard deviation values associated
        to each profile measurements
    deg_freedom : int
        number of degrees of freedom (equals
        the size of `oprofile` minus the number
        of parameters used to compute `mprofile`)
    
    """
    return (
        1. / deg_freedom *
        np.sum(np.power(oprofile - mprofile, 2) /
               np.power(ostd, 2),
               axis=0)
    )

    
def compute_likelihood(chi2_r, k, deg_freedom, log=False):
    """
    Compute the likelihood function from
    reduced chi-square.
    
    Parameters
    ----------
    chi2_r : float or array_like
        reduced chi-square values
    k : float
        a normalizing constant
    deg_freedom : int
        number of degrees of freedom (equals
        the size of `oprofile` minus the number
        of parameters used to compute `mprofile`)
    log : bool
         if True, returns the log likelihood

    """
    likelihood = k * np.exp(- deg_freedom / 2. * chi2_r)
    
    if log:
        return np.log(likelihood)
    else:
        return likelihood 


def compute_ppd(likelihood, prior):
    """
    Compute the (non-normalized) posterior
    probability distribution given the likelihood
    and the prior probability distribution.
    
    Parameters
    ----------
    likelihood : float or array_like
        the likelihood function
    prior : float or array_like
        the prior probability distribution
    
    """
    return prior * likelihood
```

### Searching (sampling) the parameter space

$P(\mathbf{m} | \mathbf{C^o_z})$ is a continuous function in the multidimensional parameter space ${\cal M}$; it is obviously impossible to calculate its value at every point (i.e., every data model). In most cases the function is also too complex to be analytically characterized. Therefore, we have to sample the parameter space, ideally with a greater number of samples in regions of high $P(\mathbf{C^o_z} | \mathbf{m})$ values (hence the 'search' term).

Here we use the *Grid Search* method, which is the most simple searching technique. It barely consists of exhaustive, uniform sampling on a regular grid that covers (a subset of) the multidimensional parameter space. We keep in mind that there exist more advanced, self-adaptive sampling techniques (e.g., SIR method, Markov Chain Monte Carlo methods) that are computationally and memory efficient, and which produce sample distributions that are more closely related to the distribution of the likehood function. Despite the lack of precision and accuracy due to its fixed and limited resolution, uniform sampling has the advantage of easily providing a global view of the parameter space. Its implementation is also straightforward, as well as the computation of marginal posterior probabilities (see below).

A Python / Numpy function to create a regular grid:

```python
def create_regular_grid(*ranges):
    """
    Returns a regular grid for uniform sampling
    in the multidimensional parameter space.
    
    Parameters
    ----------
    *ranges : range, range, ...
        parameters ranges.
        
        range can be either a `slice` object 
        or a (start, end, step) 3-tuple.
        if a complex number is be given as step,
        its real part will be then interpreted as
        the number of points to sample within the
        range.
    
    Returns
    -------
    [n-d array, n-d array, ...]
        an array of grid coordinates for each parameter.
        all arrays can be broadcasted to the regular
        grid formed by all parameters.

    See Also
    --------
    :func:`numpy.ogrid`

    """
    p_slices = [r if type(r) is slice else slice(*r)
                for r in ranges]
    
    return np.ogrid[p_slices]
   
```

### Characterizing the PPD: moments, covariance matrix and marginal PPD

Even sampled on a grid, the PPD $P(\mathbf{m} | \mathbf{C^o_z})$ is a multidimensional function that is not easy to characterize. It has, however, a few properties that are very useful for interpreting the results:

- *mean of PPD*: if the PPD is Gaussian, then the mean of the PPD equals the maximum PPD value and thus correspond to the best fit data model
- *convariance matrix of PPD*: diagonals are the posterior variances and off-diagonal terms contain information on trade-off between the model parameters.
- *1D, 2D marginal PPDs*: allow to characterize the PPD for one or two parameters with all other parameters taken into account.

The mean of the PPD for the $i$th parameter is given by:

$$ \bar{m_i} = \int_{\cal M} m_i \, P(\mathbf{m} | \mathbf{C^o_z}) \, d\mathbf{m} $$

The element of the covariance matrix corresponding to the $i$th and $j$th parameters is given by:

$$ C^M_{i,j} = \int_{\cal M} m_i m_j \, P(\mathbf{m} | \mathbf{C^o_z}) \, d\mathbf{m} - \bar{m_i} \bar{m_i} $$

The 1D marginal PPD for the $i$th parameter is given by:

$$ M(m_i) = \int ... \int P(\mathbf{m} | \mathbf{C^o_z}) \, \prod_{k=1, k \neq i}^n dm_k $$

Using a regular grid, the estimation of the integrals above is quite trivial to implement:

$$ \int_{\cal M} f(\mathbf{m}) \, d\mathbf{m} \approx \sum_g \, f(g) \, V $$

where $f$ is any function of the parameter space and $V$ is the fixed, $n$-dimensional hypervolume of each grid-box $g$.

Before evaluating the PPD properties mentionned above, we compute the integral of PPD over the grid to obtain the normalized PPD. Note that this can lead to incorrect results if the extent of the grid is not large enough to capture all the variability of the PPD (i.e., the integral of the 'true' normalized PPD over the grid would be < $\sim$1).

```python
def integrate_over_grid(grid_step, F, axis=None):
    """
    Integrate a function over a regular grid.
    
    Parameters
    ----------
    grid_step : 1-d array_like
        the resolution (step length) of one
        (`axis`) or each (`axis=None`)
        dimension of the regular grid
    F : array_like
        any function to integrate, already
        evaluated on the nodes of the regular grid
    axis : int or None
        if None, integrate over the entire grid,
        otherwise integrate only over the
        specified axis
    
    Returns
    -------
    float or n-d array
        depending on `axis`, one or several
        integrals
    
    """
    if axis is None:
        V = np.prod(grid_step)
    else:
        V = grid_step
    
    return np.sum(F * V, axis=axis)


def normalize_ppd(ppd, grid_steps):
    """
    Normalize the PPD values given on
    a regular grid, so that the integral
    over the grid equals 1.
    """
    norm = integrate_over_grid(grid_steps, ppd)
    return norm, ppd / norm 


def compute_ppd_mean_1d(ppd, grid, grid_steps, dim):
    """
    Compute the mean of the PPD for a single
    parameter.
    
    Parameters
    ----------
    ppd : array_like
        the (normalized) PPD
    grid : array_like
        grid coordinates, as returned by
        :func:`create_regular_grid`
    grid_steps : 1-d array_like
        the resolution (step length) of each
        dimension of the regular grid
    dim : int
        grid dimension number (i.e.,
        the parameter) for which the mean
        will be calculated
    
    """
    return integrate_over_grid(grid_steps, ppd * grid[dim])


def compute_ppd_covmat(ppd, grid, grid_steps):
    """
    Compute the covariance matrix of PPD.
    
    """
    dimensions = range(len(grid))
    
    ppd_mean = [compute_ppd_mean_1d(ppd, grid, grid_steps, dim)
                for dim in dimensions]
    
    CM = [[integrate_over_grid(grid_steps, ppd *
                               grid[idim] * grid[jdim])
           - ppd_mean[idim] * ppd_mean[jdim]
           for jdim in dimensions]
          for idim in dimensions]
    
    return np.array(CM)


def compute_ppd_corrmat(covmat):
    """
    Compute the correlation matrix given
    the covariance matrix `covmat`.
    """
    CrM = [[covmat[i][j] / np.sqrt(covmat[i][i] * covmat[j][j])
            for j in range(covmat.shape[1])]
           for i in range(covmat.shape[0])]
    
    return np.array(CrM)


def compute_marginal_ppd_1d(ppd, grid_steps, dim):
    """
    Compute the marginal PPD for a single
    parameter given by `dim`.
    """
    M = ppd.copy()
    ax = 0
    
    for d in range(len(grid_steps)):
        if d == dim:
            ax = -1
            continue
        M = integrate_over_grid(grid_steps[d], M, axis=ax)
    
    return M


def compute_marginal_ppd_2d(ppd, grid_steps, idim, jdim):
    """
    Compute the joint marginal PPD for two
    parameters given by `idim` and `jdim`.
    
    Returns a 2-d array
    1st dimension (rows) is the parameter i
    2st dimension (columns) is the parameter j
    
    """
    M = ppd.copy()
    ax = 0
    
    for d in range(len(grid_steps)):
        if d == idim or d == jdim:
            ax += 1
            continue
        M = integrate_over_grid(grid_steps[d], M, axis=ax)
    
    return M
```

## 3. General implementation

For more readability and usability, here is a Python class for sloving our inverse problem. It is based on the functions defined above and it allows to:

- import a measured profile data (concentrations, standard deviations...)
- define a mathematical model
- set the 'well-known' parameters and the parameters to fit
- set the sampling regular grid
- compute the modeled profiles, the likelihood function and the PPD on the sampling grid
- compute the best data model (PPD mean) the 1D and 2D marginal PPDs and the PPD covariance matrix.

It also allows to use the profiles of several nucleides in the bayesian inference.

```python
from collections import OrderedDict


class CosmogenicProfileBayesGC():
    """
    Infer a set of parameters from measured cosmogenic
    profile(s) using the Bayesian approach with the grid
    search sampling method.
    
    Parameters
    ----------
    description : string
        brief description 
    
    """
    def __init__(self, description=''):
        
        self.description = description
        self.oprofile = dict()
        self.parameters = OrderedDict()
        self.grid = None
        self.grid_size = None
        self.grid_steps = None
        self.mprofiles = None
        self.chi2_r = None
        self.likelihood = None
        self.ppd = None
        self.ppd_norm = None
        self.ppd_mean = None
        self.ppd_mean_f = None
        self.ppd_max = None
        self.ppd_max_i = None
        self.ppd_max_f = None
        self.ppd_covmat = None
        self.ppd_corrmat = None
        self.M_ppds_1d = None
    
    def set_profile_measured(self, depth, C, std, nucleide,
                             **kwargs):
        """
        Set the measured nucleide concentration profile.
        
        Parameters
        ----------
        depth : 1-d array_like
            the depth values
        C : 1-d array_like
            the measured nucleide concentration
            values
        std : 1-d array_like
            the standard deviation of the measured
            concentrations
        nucleide : 1-d array_like
            allow to distinguish concatenated profiles
            of multiple nucleides
        **kwargs : name=value, name=value...
            any other information to provide about
            the profile
        
        """
        self.oprofile['depth'] = np.array(depth)
        self.oprofile['C'] = np.array(C)
        self.oprofile['std'] = np.array(std)
        self.oprofile['nucleide'] = np.array(nucleide)
        self.oprofile.update(kwargs)
        
    def set_profile_model(self, func):
        """
        Set the mathematical model for predicting
        the comsogenic concentration profiles.

        Parameters
        ----------
        func : callable
            must accept depth values as the first
            argument and parameter value(s) as the
            other arguments for each parameter to fit,
            defined in the SAME ORDER than
            :attr:`CosmogenicProfileBayesGC.parameters` !
        
        """
        self.profile_model = func

    def set_parameter(self, name, srange, prior, **kwargs):
        """
        Set a model parameter to fit.

        Parameters
        ----------
        name : string
            name of the parameter
        srange : (start, stop, step)
            parameter search range used to compute
            the sampling regular grid. if a complex
            number is given for `step`, its real part
            will be the number of samples to generate
            instead of a step length.
        prior : callable
            the prior density probability function
            for the parameter (must accept a 1-d
            array_like as unique argument)
        **kwargs : name=value, name=value...
            any other information to provide about
            the parameter
        
        """
        p = dict()
        p['range'] = srange
        p['prior'] = prior
        p.update(kwargs)
        self.parameters[name] = p
    
    @property
    def deg_freedom(self):
        try:
            return self.oprofile['C'].size - len(self.parameters)
        except Exception:
            return None

    def _set_sampling_grid(self):
        """
        Create the sampling regular grid.
        """
        ranges = [p['range'] for p in self.parameters.values()]
        
        self.grid = create_regular_grid(*ranges)
        self.grid_sizes = [a.size for a in self.grid]
        self.grid_total_size = np.prod(self.grid_sizes)

        self.grid_steps = [1. * (stop - start) / step
                           if isinstance(step, complex)
                           else step
                           for start, stop, step in ranges]
    
    def compute_from_data_model(self, data_model):
        """
        Get modelled profiles, chi2_r, likelihood, prior
        and ppd from the given `data_model`.
        
        Returns a dictionary with the computed values.
        
        """
        f_names = ['mprofile', 'chi2_r', 'likelihood',
                   'prior', 'ppd']
        
        mprofile = self.profile_model(self.oprofile['depth'],
                                      *data_model)

        chi2_r = compute_chi2_reduced(mprofile,
                                      self.oprofile['C'],
                                      self.oprofile['std'],
                                      self.deg_freedom)
        
        likelihood = compute_likelihood(chi2_r,
                                        1.,
                                        self.deg_freedom)
        
        prior_funcs = [p['prior'] for p in self.parameters.values()]
        prior = np.prod([pf(dm)
                         for pf, dm in zip(prior_funcs, data_model)])

        ppd = compute_ppd(prior, likelihood)
        ppd /= self.ppd_norm
        
        results = dict(
            zip(f_names, [mprofile, chi2_r, likelihood, prior, ppd])
        )
        
        return results
        
    
    def compute_bayes(self, save_mprofiles=False,
                      save_chi2_r=False, save_likelihood=False):
        """
        Compute the normalized PPD, its mean,
        its covariance matrix and all the 1-d and
        2-d marginal PPDs (may take a while to compute
        and may consume a lot of memory, depending
        on the size of the sampling grid!!).
        
        The specified keyword arguments can be used to save
        the intermediate results in the corresponding
        attributes
        """
        
        if self.grid is None:
            self._set_sampling_grid()
        
        # The following operations consist of computing
        # values for all nodes of the sampling regular grid 
        
        # compute the prior distribution 
        prior_funcs = [p['prior'] for p in self.parameters.values()]
        prior = np.prod([pf(pg) for pf, pg in zip(prior_funcs, self.grid)],
                        axis=0)
        
        # array broadcasting for computing the profiles
        # for all data models (nodes of the regular grid)
        depth = self.oprofile['depth'].copy()
        oprofile = self.oprofile['C'].copy()
        ostd = self.oprofile['std'].copy()
        
        for dim in range(len(self.grid)):
            depth = np.expand_dims(depth, axis=-1)
            oprofile = np.expand_dims(oprofile, axis=-1)
            ostd = np.expand_dims(ostd, axis=-1)
        
        grid = [np.expand_dims(p, axis=0) for p in self.grid]
        
        # compute the modelled concentration profiles
        mprofiles = self.profile_model(depth, *grid)
        
        # compute the likelihood function
        chi2_r = compute_chi2_reduced(mprofiles, oprofile, ostd,
                                      self.deg_freedom)
        likelihood = compute_likelihood(chi2_r, 1., self.deg_freedom,
                                        log=False)
        
        # compute and normalize the PPD
        ppd = compute_ppd(likelihood, prior)
        self.ppd_norm, self.ppd = normalize_ppd(ppd, self.grid_steps)
        
        # keep or delete intermediate results
        if save_mprofiles:
            self.mprofiles = mprofiles
        else:
            del mprofiles
        if save_chi2_r:
            self.chi2_r = chi2_r
        else:
            del chi2_r
        if save_likelihood:
            self.likelihood = likelihood
        else:
            del likelihood
        
        del ppd
        del grid
        
        # compute PPD mean and mode (+ functions values)
        self.ppd_mean = [compute_ppd_mean_1d(self.ppd, self.grid,
                                              self.grid_steps, dim)
                         for dim in range(len(self.grid))]
        
        self.ppd_mean_f = self.compute_from_data_model(self.ppd_mean)
        
        self.ppd_max_i = np.nonzero(self.ppd >= self.ppd.max())
        self.ppd_max = [np.arange(*p['range'])[m][0]
                         for p, m in zip(self.parameters.values(),
                                         self.ppd_max_i)]

        self.ppd_max_f = self.compute_from_data_model(self.ppd_max)
        
        # compute PPD covavriance and correlation matrices
        self.ppd_covmat = compute_ppd_covmat(self.ppd,
                                             self.grid,
                                             self.grid_steps)
        
        self.ppd_corrmat = compute_ppd_corrmat(self.ppd_covmat)
        
        
        # compute 1D marginal PPDs and find maximums
        self.M_ppds_1d = [compute_marginal_ppd_1d(self.ppd,
                                                  self.grid_steps,
                                                  dim)
                          for dim in range(len(self.grid))]
        
        M_ppds_1d_max_i = [M.argmax()
                           for M in self.M_ppds_1d]
        self.M_ppds_1d_max = [
            np.arange(*p['range'])[m]
            for p, m in zip(cosmoinv.parameters.values(),
                            M_ppds_1d_max_i)
        ]
        
        self.M_ppds_1d_max_f = self.compute_from_data_model(
            self.M_ppds_1d_max
        )
        
        # compute 2D marginal PPDs
        self.M_ppds_2d = [[compute_marginal_ppd_2d(self.ppd,
                                                  self.grid_steps,
                                                  idim, jdim)
                           for jdim in range(len(self.grid))]
                          for idim in range(len(self.grid))]
    
    
    def setup_summary(self):
        if self.grid is None:
            self._set_sampling_grid()
        
        summary = "Modelling C profile (Bayes, Grid-Search)\n\n"
        summary += "DESCRIPTION:\n{desc}\n\n".format(
                       desc=self.description
                   )
        summary += "MEASURED PROFILE ({N} samples):\n".format(
                       N=len(self.oprofile['C'])
                   )
        summary += str(pd.DataFrame(self.oprofile))
        summary += "\n\n"
        summary += "PROFILE MODEL:\n{fname}\n{fdoc}\n\n".format(
                       fname=self.profile_model.__name__,
                       fdoc=inspect.getdoc(self.profile_model)
                   )
        summary += "'UNKNOWN' PARAMETERS ({n}):\n".format(
                       n=len(self.parameters)
                   )
        summary += "\n".join([
                       name + ":\n" +
                       "\n".join(["\t{0}: {1}".format(k, v)
                                  for k, v in p.items()])
                        for name, p in self.parameters.items()
                   ])
        summary += "\n\ndegrees of freedom: {dof}\n\n".format(
                       dof=self.deg_freedom
                   )
        summary += "GRID SEARCH:\n"
        summary += "nb. of nodes per parameter: {np}\n".format(
                       np=self.grid_sizes
                   )
        summary += "total nb. of nodes: {ng}\n\n".format(
                       ng=self.grid_total_size
                   )
        
        return summary
    
    def results_summary(self):
        summary = "RESULTS:\n\n"
        
        if self.ppd is None:
            return summary + "no result yet"
        
        summary += "parameter names in order:\n{0}\n\n".format(
                       self.parameters.keys()
                   )
        summary += "PPD max:\n{0}\n\n".format(self.ppd_max)
        summary += "Values at PPD max:\n"
        summary += "\n".join(["{0}:\n {1}".format(k, v)
                              for k, v in self.ppd_max_f.items()])
        summary += "\n\n"
        summary += "PPD mean:\n{0}\n\n".format(self.ppd_mean)
        summary += "Values at PPD mean:\n"
        summary += "\n".join(["{0}:\n {1}".format(k, v)
                              for k, v in self.ppd_mean_f.items()])
        summary += "\n\n"
        summary += "1D Marginal PPD maxs:\n{0}\n\n".format(
                        self.M_ppds_1d_max
                   )
        summary += "Values at 1D Marginal PPD maxs:\n"
        summary += "\n".join(["{0}:\n {1}".format(k, v)
                              for k, v in self.M_ppds_1d_max_f.items()])
        summary += "\n\n"
        summary += "PPD covmat:\n{0}\n\n".format(self.ppd_covmat)
        summary += "PPD corrmat:\n{0}\n\n".format(self.ppd_corrmat)
        
        return summary
    
    def __str__(self):
        return self.setup_summary() + self.results_summary()
```

### Mathematical Model

- We set the mathematical model and all its fixed "known" parameters. The function must accept profile depth as its first argument, "unknown" parameters as other arguments and optionally keyword arguments.

```python
def C_10Be(depth, erosion, exposure, density, inheritance):
    """
    A model for 10Be nucleide concentration profiles
    with 4 unknowns (erosion rate, exposure time,
    soil density and inheritance)
    
    """
    
    # nucleide parameters
    berillium10 = {
        'rdecay': math.log(2.) / 1.36e6
    }
    
    # particles parameters
    neutrons = {
        'prod_rate': 0.9785 * site_settings['P_0'],
        'damping_depth': 160.
    }
    
    slow_muons = {
        'prod_rate': 0.015 * site_settings['P_0'],
        'damping_depth': 1500.
    }
    
    fast_muons = {
        'prod_rate': 0.0065 * site_settings['P_0'],
        'damping_depth': 5300.
    }
    
    # other, fixed parameters
    #density = 1.9
    #inheritance = 2.5e5
    
    return compute_Cn(depth, erosion, exposure,
                      density, inheritance,
                      berillium10,
                      [neutrons, slow_muons, fast_muons])
```

### Bayesian / Grid Search setup

- Create a new object for setup and results

```python
cosmoinv = CosmogenicProfileBayesGC(description='Belle Roche 10Be 4-params')
```

- Add the measurements

```python
cosmoinv.set_profile_measured(
    profile_data['depth_cm'].values,
    profile_data['C_'].values,
    profile_data['std'].values,
    profile_data['nucleide'].values,
    depth_gcm_2=profile_data['depth_g-cm-2'].values
)
```

- Add the mathematical model

```python
cosmoinv.set_profile_model(C_10Be)
```

- Set the unknown parameters, their prior and their search range. **CAUTION**: The parameters must be defined in the same order than the arguments in the mathematical model function!!

```python
cosmoinv.set_parameter(
    'erosion_rate',
    [0, 7e-4, 0.5e-5],
    stats.uniform(loc=0, scale=7e-4).pdf
    #[0, 5e-4, 1e-5],
    #stats.uniform(loc=0, scale=5e-4).pdf
)

cosmoinv.set_parameter(
    'exposure_time',
    [0, 6e5, 5e3],
    stats.uniform(loc=0, scale=6e5).pdf
    #[0.0, 5e5, 5e3],
    #stats.uniform(loc=0, scale=5e5).pdf
)

cosmoinv.set_parameter(
    'soil_density',
    [1.8, 2.8, 0.04],
    stats.uniform(loc=1.8, scale=2.8).pdf
    #[1.8, 2.4, 0.1],                          # from Rixhon et. al.: poor resolution
    #stats.uniform(loc=1.8, scale=2.4).pdf
)

cosmoinv.set_parameter(
    'inheritance',
    [0, 1e5, 0.2e4],
    stats.uniform(loc=0, scale=1e5).pdf
    #[0, 1e5, 1e4],                             # from Rixon et al.: poor resolution
    #stats.uniform(loc=0, scale=1e5).pdf
)
```

## Plot results

- Plot 1D marginal PPDs (blue lines) and prior probability distributions (red lines)

```python
%matplotlib inline

def plot_marginal_1d(cbayes, pname, ax=None):
    
    p = cbayes.parameters[pname]
    pindex = cbayes.parameters.keys().index(pname)
    
    x = cbayes.grid[pindex].flatten()
    Mprior = p['prior'](x)
    Mposterior = cbayes.M_ppds_1d[pindex]
    
    ymax = max(Mprior.max(), Mposterior.max())

    if ax is None:
        ax = plt.subplot(111)
    
    ax.plot(x, Mposterior, label='posterior')
    ax.fill_between(x, Mposterior, alpha=.65)
    ax.plot(x, Mprior, 'r-', label='prior')
    
    plt.setp(ax, xlabel=pname,
             yticklabels=[],
             xlim=p['range'][0:2],
             ylim=[0, ymax])
    

def plot_marginals_1d(cbayes, n_subplot_cols=2, **kwargs):
    
    n_subplots = len(cbayes.parameters)
    n_subplot_rows = int(math.ceil(1. * 
                                   n_subplots /
                                   n_subplot_cols))
    
    fig, aax = plt.subplots(nrows=n_subplot_rows,
                            ncols=n_subplot_cols,
                            **kwargs)
    axes = aax.flatten()
    fig.text(0.5, 0.975,
             "Marginal PPDs: " + cbayes.description,
             horizontalalignment='center',
             verticalalignment='top')
    
    for i, pname in enumerate(cbayes.parameters.keys()):
        ax = axes[i]
        plot_marginal_1d(cbayes, pname, ax=ax)
    
    plt.tight_layout()
    plt.subplots_adjust(top=0.93)


plot_marginals_1d(cosmoinv, figsize=(8, 6))
```

- Show 2D marginal PDDs 

```python
def plot_marginal_2d(cbayes, p1p2, ax=None,
                     cmap='Blues', show_colorbar=False):
    
    pname1, pname2 = p1p2
    idim = cbayes.parameters.keys().index(pname2)
    jdim = cbayes.parameters.keys().index(pname1)
    
    if ax is None:
        ax = plt.subplot(111)
    
    X, Y = np.meshgrid(cbayes.grid[idim].flatten(),
                       cbayes.grid[jdim].flatten())
    
    M2D = ax.contourf(Y, X,
                      cbayes.M_ppds_2d[idim][jdim],
                      cmap=plt.get_cmap(cmap))
    
    plt.setp(ax, xlabel=pname1, ylabel=pname2)
    
    if show_colorbar:
        plt.colorbar(M2D, ax=ax)
    

fig = plt.figure(figsize=(9, 9))
ax = plt.subplot(321)
plot_marginal_2d(cosmoinv, ('erosion_rate', 'exposure_time'), ax=ax)
ax2 = plt.subplot(322)
plot_marginal_2d(cosmoinv, ('exposure_time', 'inheritance'), ax=ax2)
ax3 = plt.subplot(323)
plot_marginal_2d(cosmoinv, ('erosion_rate', 'inheritance'), ax=ax3)
ax4 = plt.subplot(324)
plot_marginal_2d(cosmoinv, ('exposure_time', 'soil_density'), ax=ax4)
ax5 = plt.subplot(325)
plot_marginal_2d(cosmoinv, ('erosion_rate', 'soil_density'), ax=ax5)
ax6 = plt.subplot(326)
plot_marginal_2d(cosmoinv, ('soil_density', 'inheritance'), ax=ax6)
plt.tight_layout()
```

- Show PPD correlation matrix

```python
def plot_coormat(cbayes):
    ax = plt.subplot(111)
    
    CrM = ax.pcolormesh(cbayes.ppd_corrmat, 
                        cmap=plt.get_cmap('coolwarm'),
                        vmin=-1, vmax=1)
    
    n = np.arange(0.5, len(cbayes.parameters))
    labels = cbayes.parameters.keys()
    plt.xticks(n, labels ,rotation='vertical')
    plt.yticks(n, labels)
    
    plt.grid(False)
    plt.colorbar(CrM, ax=ax)
    
    ax.set_title("PPD Correlation matrix: " + cbayes.description)
    plt.subplots_adjust(top=0.9)


fig = plt.figure(figsize=(5, 5))
plot_coormat(cosmoinv)
```

- Plot measured and modelled profiles

```python
def plot_profiles(cbayes, *args, **kwargs):
    ax = plt.subplot(111)
    ax.errorbar(cbayes.oprofile['C'], -1. * cbayes.oprofile['depth'],
                xerr=cbayes.oprofile['std'],
                fmt='o',
                label='measured')
    
    show_bayes = kwargs.get('show_bayes') or True
    
    if show_bayes:
        ax.plot(C_10Be(cbayes.oprofile['depth'], *cbayes.ppd_max),
                -1. * cbayes.oprofile['depth'],
                label='PPD mode')

        ax.plot(C_10Be(cbayes.oprofile['depth'], *cbayes.ppd_mean),
                -1. * cbayes.oprofile['depth'],
                label='PPD mean')

        ax.plot(C_10Be(cbayes.oprofile['depth'], *cbayes.M_ppds_1d_max),
                -1. * cbayes.oprofile['depth'],
                label='1D marginal PPDs maxs')
    
    for label, data_model in args:
        ax.plot(C_10Be(cbayes.oprofile['depth'], *data_model),
                -1. * cbayes.oprofile['depth'],
                label=label)
    
    plt.setp(ax, xlabel='nucleide concentration [atoms g-1]',
             ylabel='depth [cm]')
    ax.legend(loc='lower right')
    ax.set_title("Profiles: " + cbayes.description)


fig = plt.figure(figsize=(7, 5))
plot_profiles(cosmoinv, ('Rixhon et al. 2011', [4e-4, 222.5e3, 2.3, 20e3]))
```
