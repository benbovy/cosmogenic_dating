## Mathematical (Predictive) Model

We implement the general equations used for modelling cosmogenic nucleides concentration profiles

```python
def particle_contrib(depth, erosion, exposure, density,
                     nucleide, particle):
    """
    Returns the contribution of a specific particle type
    to the nucleide concentration.
    
    See Also
    --------
    C_nucleide
    """
    return (
        particle['prod_rate'] / ((density * erosion / 
                                  particle['damping_depth']) 
                                 + nucleide['rdecay']) *
        np.exp(-1. * density * depth / particle['damping_depth']) *
        (1. - np.exp(-1. * exposure * ((erosion * density / 
                                       particle['damping_depth'])
                                       + nucleide['rdecay'])))
    )


def C_nucleide(depth, erosion, exposure, density,
               inheritance, nucleide, particles):
    """
    Calculate the concentration(s) of a nucleide
    at given depth(s) (generic function).
    
    Parameters
    ----------
    depth : float or array_like
        depth(s) below the surface [cm]
    erosion : float or array_like
        erosion rate(s) [cm yr-1]
    exposure : float or array_like
        exposure time [yr]
    density : float or array_like
        soil density [g cm-3]
    inheritance : float or array_like
        concentration of the nucleide at the
        initiation of the exposure scenario
        [atoms g-1]
    nucleide : dict
        nucleide parameters
    particles : [dict, dict, ...]
        parameters related to each particule type
        that contribute to the nucleide production
    
    Returns
    -------
    float or array-like (broadcasted)
        the nucleide concentration(s) [atoms g-1]
    
    Notes
    -----
    if arrays are given for several arguments, they must
    have the same shape or must be at least be broadcastable.
    
    `nucleide` must have the following key(s):
        'rdecay': radioactive decay [yr-1]
    
    each item in `particles` must have the following keys:
        'prod_rate': surface production rate [atoms g-1 yr-1]
        'damping_depth': effective apparent attenuation depth
                         [g cm-2]
    
    """
    return (
        np.sum((particle_contrib(depth, erosion, 
                                 exposure, density,
                                 nucleide, p)
                for p in particles),
               axis=0) +
        inheritance * np.exp(-1. * nucleide['rdecay'] * exposure)
    )


def C_10Be(depth, erosion, exposure, density, inheritance,
           P_0=5.3):
    """
    A model for 10Be nucleide concentration profiles
    with 4 unknowns: erosion rate, exposure time,
    soil density and inheritance.
    
    Notes
    -----
    The following parameters are used:
    
    10Be radioactive decay: log 2 / 1.36e6
    
    Contribution of
    - neutrons
        - production rate: 0.9785 * P_0
        - damping depth: 160
    - slow muons
        - production rate: 0.015 * P_0
        - damping depth: 1500
    - fast muons
        - production rate: 0.0065 * P_0
        - damping depth: 5300
    
    See Also
    --------
    C_nucleide
    
    """
    
    # nucleide parameters
    berillium10 = {
        'rdecay': math.log(2.) / 1.36e6
    }
    
    # particles parameters
    neutrons = {
        'prod_rate': 0.9785 * P_0,
        'damping_depth': 160.
    }
    
    slow_muons = {
        'prod_rate': 0.015 * P_0,
        'damping_depth': 1500.
    }
    
    fast_muons = {
        'prod_rate': 0.0065 * P_0,
        'damping_depth': 5300.
    }
    
    return C_nucleide(depth, erosion, exposure,
                      density, inheritance,
                      berillium10,
                      [neutrons, slow_muons, fast_muons])


def 

def residuals(params, model, oprofile,
              params_map, params_fixed, 
              mprofile=None):
    """
    Compute the normalised residuals
    """
    if mprofile is None:
        p = [params[pm] if pm is not None else pf
             for pm, pf in zip(params_map, params_fixed)]
        
        mprofile = model(oprofile['depth'], *p)
    
    return (oprofile['C'] - mprofile) / oprofile['std']


def chi2(params, model, oprofile, mprofile=None):
    """
    Compute the Chi-square value 
    """
    if mprofile is None:
        mprofile = model(oprofile['depth'], *params)
    
    return (
        np.sum(np.power(oprofile['C'] - mprofile, 2) /
        np.power(oprofile['std'], 2),
        axis=0)
    )


def loglikelihood(params, model, oprofile, mprofile=None):
    """
    Compute the log-likelihood
    """
    if mprofile is None:
        mprofile = model(oprofile['depth'], *params)
    
    std_square = np.power(oprofile['std'], 2)
    
    return -0.5 * (
        np.sum(np.power(oprofile['C'] - mprofile, 2) /
               inv_sigma2 - np.log(2 * np.pi * std_square),
               axis=0)
    )
    

class CosmogenicProfileMLE():
    """
    Infer a set of parameters from measured cosmogenic
    profile(s) using Maximum Likelihood Estimation.
    
    Parameters
    ----------
    description : string
        brief description 
    
    """
    def __init__(self, description=''):
        
        self.description = description
        self.oprofile = dict()
        self.parameters = OrderedDict()
        self.grid = None
        self.grid_size = None
        self.grid_steps = None
    
    def set_profile_measured(self, df):
        """
        Set the measured nucleide concentration profile.
        
        Parameters
        ----------
        df : :class:`pandas.DataFrame` object
            a dataframe object with at least the
            following columns: 'depth', 'C', 'std' and
            'nucleide'
        """
        self.oprofile = df
        
    def set_profile_model(self, C_func, pmap, **kwargs):
        """
        Set the mathematical model for predicting
        the comsogenic concentration profiles.

        Parameters
        ----------
        C_func : callable
            must accept depth values as the first
            argument and parameter value(s) as the
            other arguments for each parameter to fit
        pmap : list
            list of parameter names to map to the
            parameter arguments of `C_func`
        **kwargs :
            keyword arguments to pass to `C_func`
        
        """
        self.mprofile_func = lambda depth, *p: C_func(depth, *pmap, **kwargs)

    def set_parameter(self, name, srange, prior, **kwargs):
        """
        Set a model parameter to fit.

        Parameters
        ----------
        name : string
            name of the parameter
        srange : (start, stop, step)
            parameter search range used to compute
            the sampling regular grid. if a complex
            number is given for `step`, its real part
            will be the number of samples to generate
            instead of a step length.
        prior : callable
            the prior density probability function
            for the parameter (must accept a 1-d
            array_like as unique argument)
        **kwargs : name=value, name=value...
            any other information to provide about
            the parameter
        
        """
        p = dict()
        p['range'] = srange
        p['prior'] = prior
        p.update(kwargs)
        self.parameters[name] = p
    
    @property
    def deg_freedom(self):
        return self.oprofile['C'].size - len(self.parameters)

    def _set_sampling_grid(self):
        """
        Create the sampling regular grid.
        """
        ranges = [p['range'] for p in self.parameters.values()]
        
        self.grid = create_regular_grid(*ranges)
        self.grid_sizes = [a.size for a in self.grid]
        self.grid_total_size = np.prod(self.grid_sizes)

        self.grid_steps = [1. * (stop - start) / step
                           if isinstance(step, complex)
                           else step
                           for start, stop, step in ranges]
```

## Data

Test case

```python
true_erosion = 2e-3
true_exposure = 1.5e5
true_density = 2.0
true_inheritance = 0.

m_error_C_ratio = 1./10
n_samples = 10

profile_data = pd.DataFrame()

profile_data['depth_cm'] = np.linspace(50, 1000, n_samples)
profile_data['C_'] = C_10Be(profile_data['depth_cm'],
                            true_erosion,
                            true_exposure,
                            true_density,
                            true_inheritance)
profile_data['std'] = profile_data['C_'] * m_error_C_ratio

noise = np.array([np.random.normal(scale=C * m_error_C_ratio)
                  for C in profile_data['C_']]).flatten()
profile_data['C_'] += noise

profile_data['nucleide'] = np.repeat('10Be', n_samples)

profile_data
```

## Fit the model to data

```python
def residuals(params):
    """
    
    
    """
    erosion = params['erosion_rate'].value
    exposure = params['exposure_time'].value
    density = params['soil_density'].value
    inheritance = params['inheritance'].value
    
    C_mod = C_10Be(profile_data['depth_cm'],
                   erosion, exposure, density, inheritance,
                   P_0=5.3)
    
    return (C_mod - profile_data['C_']) / profile_data['std']
    
```

```python
params = lmfit.Parameters()
params.add('erosion_rate',   value=5e-4, min=0.)
params.add('exposure_time', value=1e5, min=0.)
params.add('soil_density', value=2.3)
params.add('inheritance', value=5e3, min=0.)
```

```python
result = lmfit.minimize(residuals, params, ftol=1e-7, xtol=1e-7)
```
